 ### DATA PREPARATION

Since our columns contains missing values and Null values, we will clean our datatset by filling the missing values.

```python
# check sum of null values
df.isnull().sum()
```

```python
#drop columns not needed 
drop_cols = ['Airport.Code', 'Aircraft.Category', 'Amateur.Built', 'Airport.Name', 'Registration.Number', 'FAR.Description', 'Schedule', 'Publication.Date', 'Purpose.of.flight', 'Latitude', 'Longitude', 'Air.carrier', 'Broad.phase.of.flight']
df.drop(columns = drop_cols, inplace = True)
```

```python
#columns needed
df.columns
```

```python
#Drop rows with missing Make or Model
df.dropna(subset=['Make', 'Model', 'Report.Status'], inplace=True)
df
```

```python
df["State"] = df["Location"].str[-2:]
```

```python
# Normalize and clean Location
df['Location'] = df['Location'].str.upper().str.strip()
df['Location'] = df['Location'].fillna('UNKNOWN')

# Normalize and clean Country
df['Country'] = df['Country'].str.upper().str.strip()
df['Country'] = df['Country'].fillna('UNKNOWN')
df
```

```python
# Extract proper 2-letter state codes
df['State'] = df['Location'].str.extract(r',\s*([A-Z]{2})$')

# Fill missing state codes with 'UNKNOWN'
df['State'] = df['State'].fillna('UNKNOWN')
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# inspect the values in no. of engines and engine types
df['Number.of.Engines'].value_counts(dropna=False)
df['Engine.Type'].value_counts(dropna=False)
df
#fill missing values with unknown
df['Number.of.Engines'] = df['Number.of.Engines'].fillna(0).astype(int)
df['Engine.Type'] = df['Engine.Type'].fillna('UNKNOWN').str.upper().str.strip()
df
```

```python
#filling weather condition with placeholder
df['Weather.Condition'] = df['Weather.Condition'].str.upper().str.strip()
df['Weather.Condition'] = df['Weather.Condition'].fillna('UNKNOWN')
df
```

```python
#filling missing with a placeholder
#Normalize the values
df['Aircraft.damage'] = df['Aircraft.damage'].str.upper().str.strip()

#Fill missing values with a placeholder
df['Aircraft.damage'] = df['Aircraft.damage'].fillna('UNKNOWN')
df
```

```python
df.isnull().sum()
```

```python
#create a variable injury columns
#convert injury columns with zeros and convert to integer
injury_cols = [
    'Total.Fatal.Injuries', 'Total.Serious.Injuries', 
    'Total.Minor.Injuries', 'Total.Uninjured'
]
df[injury_cols] = df[injury_cols].fillna(0).astype(int)
df
```

```python
df.isnull().sum()
```

```python
# fill missing data in injury.severity
df['Injury.Severity'] = df['Injury.Severity'].str.upper().str.strip()
df['Injury.Severity'] = df['Injury.Severity'].str.extract(r'(FATAL|SERIOUS|MINOR|NONE|UNKNOWN|INCIDENT|NON-FATAL)', expand=False)
df['Injury.Severity'] = df['Injury.Severity'].fillna('UNKNOWN')
df
```

### Aircraft Frequency and Manufacturing Trends

```python
df['Make'] = df['Make'].str.title()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot Top 10 makes
top_makes = df['Make'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_makes.values, y=top_makes.index)
plt.title('Top 10 Aircraft Manufacturers by Number of Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Make')
plt.show()
```

High Injury Totals correspond to High Exposure or High Risk of number of accidents.
Cessna dominates being the list having the higher relative  number of accidents folowed by Piper and Beech.
This may be due to been widely used in aviation.
Hughes being the safest manufacturer having less ralative number of accidents.


## Total Deaths by Make

```python
df.groupby(['Make'])['Total.Fatal.Injuries'].sum().head(10).plot(kind = 'bar', figsize = (10,6))
plt.title('Total Deaths by Make')
plt.xlabel('Make')
plt.ylabel('Deaths')
plt.show()
```

### Weather Condition and Risk

```python
plt.figure(figsize=(10, 6))
sns.countplot(
    data=df,
    x='Weather.Condition',
    hue='Injury.Severity',
    order=['VMC', 'IMC', 'UNKNOWN', 'UNK'],  # Adjust based on what's present
    palette='Set2'
)
plt.title('Aircraft Accident Severity by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.legend(title='Injury Severity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.subplots_adjust(top=0.88, bottom=0.18)
plt.show()
```

 
Most accidents occur in VMC (Visual Meteorological Conditions), this could be because more flights happen then.
IMC shows proportionally higher serious/fatal injuries 


## Injury Severity Distribution

## Top 10 States with Most Aircraft Accidents

```python
#states-level analysis(USA only)
state_counts = df[df['State'] != 'UNKNOWN']['State'].value_counts().head(10)
sns.barplot(x=state_counts.values, y=state_counts.index)
plt.title('Top 10 States with Most Aircraft Accidents')
plt.xlabel('Accident Count')
plt.show()
```

## Damage Severity by Engine Type

```python
engine_damage = df.groupby('Engine.Type')['Aircraft.damage'].value_counts(normalize=True).unstack()
engine_damage.plot(kind='hist', stacked=True, figsize=(8,4))
plt.title('Damage Severity by Engine Type')
plt.ylabel('Aircraft.damage')
plt.xlabel('Engine.Type')
plt.show()
```

There is higher proption of Destroyed damage in Low Engine type suggesting that certain engine complexity or characteristics are more prone to being destroyed in accidents.
Substantial damage is concentrated in Higher Engine type indicating that powerful engines tend to sustain substantial not total damage.
Minor is common in Lower engine proportions indicating either involved in less severe accidents or are more resilient to damage.

## Injury Severity Under  Different Weather Conditions

```python
weather_severity = pd.crosstab(df['Weather.Condition'], df['Injury.Severity'])
weather_severity.plot(kind='hist', stacked=False, figsize=(10,6), colormap='coolwarm')
plt.title("Injury Severity under Different Weather Conditions")
plt.xlabel("Weather Condition")
plt.ylabel("Number of Accidents")
plt.show()
```

The distribution is highly sparse and skewed.
FATAL injuries are rare but do occur under specific weather conditions. This could suggest a relationship between weather severity and injury outcomes.

The overwhelming majority of accidents occur under VMC
Most aircraft accidents occur in good weather — but that’s when most flying happens. However, poor weather conditions (IMC) are disproportionately fatal. When evaluating aircraft for purchase, models with strong instrumentation, proven stability, and IFR (Instrument Flight Rules) ratings should be prioritized to reduce weather-related risk exposure.
